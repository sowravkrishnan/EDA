### Data Story
We have observed there are null values in the column College and Salary.

```python
Q1 = des['Salary'].loc['25%']
Q3 = des['Salary'].loc['75%']
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = (df['Salary'] < lower_bound) | (df['Salary'] > upper_bound)

print(df[outliers][['Name', 'Team', 'Position', 'Salary']])
```

### Data Story
We need to identify if there are any outliers in the Salary column to determine which method we need to use to fill null values. Here we use IQR method,where we find IQR(Inter quartile range). If the values are outside this range then they are outliers.We have all the outliers listed.

```python
df['College'].fillna('Unknown',inplace=True)
df_sorted = df.sort_values(by='Salary')
df['Salary'].fillna(df_sorted['Salary'].median(),inplace=True)
df.isnull().sum()
```

### Data Story
As we found there are outliers, we have used median of the salary values and filled the null values with median. We filled the null values in college column with unknown. Then we used is null function to verify that there is no other null values.

# <p style="text-align: center;">Data Visualisation</p>

```python
Countdf=pd.DataFrame({"No_of_Employees":df.Team.value_counts()}).reset_index()
Countdf.columns = ['Team','No_of_Employees']
Countdf['Abbreviation'] = Countdf['Team'].apply(lambda x: ''.join(word[0] for word in x.split()))
Countdf = Countdf[['Team', 'Abbreviation', 'No_of_Employees']]
Countdf
```

### Data Story
We have created a dataframe Countdf which has details of Number of employees in each team. We added another column called abbreviation and printed the short form for each team. 

```python
plt.figure(figsize=(15, 10))
plt.bar(Countdf['Abbreviation'],Countdf['No_of_Employees'],color=plt.cm.tab20c.colors)
plt.title('Number of Employees in Each Team')
plt.xlabel('No of Employees')
plt.ylabel('Team')
plt.show()
```

### Data Story
We plotted a bar plot of No of Employees in each team.

```python
Countdf['Percentage']= (Countdf['No_of_Employees'] / Countdf['No_of_Employees'].sum()) * 100
Countdf
```

### Data Story
We have added another column Percentage to show the percentage contribution of each team w.r.t total number of employees.

```python
plt.pie(Countdf['Percentage'],labels=Countdf['Abbreviation'],colors=plt.cm.tab20c.colors)
plt.title('Percentage of Employees in Each Team')

plt.show()
```
